```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 
```

## Histograms


```python
img = cv2.imread('photografer.jpg',0)
hist,bins = np.histogram(img.flatten(),256,[0,256])

plt.imshow(img, cmap='gray')
plt.show()

plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.show()
```

## Min and Max

```python
print('min is',hist.min())
print('max is',hist.max())

```

## Image normalization

```python
img = cv2.imread('photografer.jpg')
norm_image = cv2.normalize(img, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
plt.hist(norm_image.flatten())
plt.show()
cv2.imwrite('normalized-photografer.jpg', norm_image)
```

```python
norm_image = cv2.imread('normalized-photografer.jpg',0)
plt.imshow(norm_image, cmap='gray')
```

## Histograms Equalization 

```python
img = cv2.imread('photografer.jpg',0)
equ = cv2.equalizeHist(img)
res = np.hstack((img,equ)) 
# save image
cv2.imwrite('photografer2.jpg',equ)
plt.hist(equ.flatten(),256,[0,256], color = 'b')
plt.show()
plt.imshow(res, cmap='gray')
```

## Valeur seuil OTSU

```python
cv2.namedWindow('Image binaire OTSU')
thresh,imgbin=cv2.threshold(img,0,255,cv2.THRESH_OTSU)
print( 'Valeur seuil OTSU ', thresh)
```
